```python
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set()
```

```python
with open('tom_1.txt', 'r', encoding='utf8') as f:
    text = f.read().lower()
with open('tom_2.txt', 'r', encoding='utf8') as f:
    text += f.read().lower()
with open('tom_3.txt', 'r', encoding='utf8') as f:
    text += f.read().lower()
with open('tom_4.txt', 'r', encoding='utf8') as f:
    text += f.read().lower()
    
text = re.sub('[^а-я]','', text)
```

```python
RUSSIAN = 'абвгдежзийклмнопрстуфхцчшщъыьэюя'
POS = {l: i for i, l in enumerate(RUSSIAN)}
```

```python
probabilities = np.zeros((len(RUSSIAN), len(RUSSIAN)))
```

```python
for cp, cn in zip(text[:-1], text[1:]):
    probabilities[POS[cp], POS[cn]] += 1
```

```python
probabilities = (probabilities / probabilities.sum(axis=1)[:, None])
```

```python
df = pd.DataFrame(probabilities, index=(c for c in RUSSIAN), columns=(c for c in RUSSIAN))
```

```python
df
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(df)
plt.show()
```

```python
proba_sum = 0
for s1 in RUSSIAN:
    for s2 in RUSSIAN:
        string = 'п' + s1 + s2 + 'р'
        proba = 1.0 / 32 #отсутсвует буква ё
        for cp, cn in zip(string[1:], string[:-1]):
            proba *= probabilities[POS[cp], POS[cn]]
        proba_sum += proba
print(proba_sum)
```

```python
s = []
start = 'б'
s.append(start)
length = 8
for i in range(length-1):
    index = POS[s[i]]
    next_char = np.random.choice(list(RUSSIAN), p=probabilities[index])
    s.append(next_char)

print (''.join(s))
```
